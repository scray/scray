```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def createBarplot(md=None,fx=24,fy=12,fontscale=3.0,title="") :
    sns.set(style='whitegrid', palette='muted', font_scale=fontscale)
    plt.figure(figsize=(fx,fy))
    plt.title(title)
    ax = sns.barplot(x=md['date'], y=md['outcome'], data=md)
    plt.setp( ax.xaxis.get_majorticklabels(), rotation=75 )
    plt.tight_layout()
    plt.show()
    return ax

## heatmap
def createHeatmap(piv,title="",vmin=None,vmax=None) :
    plt.figure(figsize=(24,8))
    plt.title(title)
    if vmin == None:
        ax = sns.heatmap(piv, cbar=True, square=True)
    else:
        ax = sns.heatmap(piv, cbar=True, square=True, vmin=vmin, vmax=vmax)
    plt.setp( ax.xaxis.get_majorticklabels(), rotation=0 )
    plt.tight_layout()
    plt.show()
    return ax
```

```python
import datetime

def label(graph,skip,rot) :
    for ind, label in enumerate(graph.get_xticklabels()):
        if ind % skip == 0:  # every 10th label is kept
            label.set_visible(True)
            label.set_rotation(rot)
        else:
            label.set_visible(False)

def get_ymdh(mdcountsall) :
    levels = []
    for i in range(mdcountsall.index.nlevels):
        levels.append(mdcountsall.index.get_level_values(i).astype(str))
    return levels

def make_2digits(blist):
    for n, b in enumerate(blist):
        if int(b) < 10:
             blist[n] = '0' + str(b)
    return blist

def get_ymd_string(a,b,c) :
    if isinstance(a, str) :
        return a + "-" + make_2digits([b])[0] + "-" + make_2digits([c])[0] 
    return a + "-" + pd.Index(make_2digits(b.tolist())) + "-" + pd.Index(make_2digits(c.tolist())) 


def get_ymdh_string(levels) :
    value = ''
    for level in levels:
        if len(value) > 1:
            value += "-"
        #value += level
        value += pd.Index(make_2digits(level.tolist()))
    return value    
    #return a + "-" + b + "-" + c + "-" + d

def aggregate(pfall,month,column='counter',groupby=['year','month','day','hour']) :
    if (month > 0) & (month < 13) :
        mdcountsall = pfall[(pfall['month'] == month)].groupby(groupby)[column].sum()
    else :
        mdcountsall = pfall.groupby(groupby)[column].sum()  
    
    levels = get_ymdh(mdcountsall)
    data2 = pd.DataFrame()
    data2['date'] = get_ymdh_string(levels)
        
    data2['year']  =  levels[0].astype(int) 
    data2['month'] =  levels[1].astype(int) 
    data2['day']   =  levels[2].astype(int)     
        
    data2['outcome'] =  mdcountsall.reset_index(level=0, drop=True).reset_index()[column].astype(int)    
    return data2.sort_values(['date']).reset_index()
        
def createData(pfall,month,column='counter',groupby=['year','month','day','hour']) :
    if (month > 0) & (month < 13) :
        mdcountsall = pfall[(pfall['month'] == month)].groupby(groupby)[column].sum()
    else :
        mdcountsall = pfall.groupby(groupby)[column].sum()     
        
    levels = get_ymdh(mdcountsall)
    data2 = pd.DataFrame()
    data2['date'] = get_ymdh_string(levels)
   
    data2['year']  =  levels[0].astype(int) 
    data2['month'] =  levels[1].astype(int) 
    data2['day']   =  levels[2].astype(int) 
    
    data2['outcome'] =  mdcountsall.reset_index(level=0, drop=True).reset_index()[column].astype(int)
        
    #for pivot table
    if 'hour' in groupby:
        data2['hours'] =  levels[3].astype(int) 
        data2['days']  =  levels[2].astype(int) 
    return data2
```

```python
class BaseAction(object):

    def from_dict(self,dictionary):
        for k, v in dictionary.items():
            setattr(self, k, v)
        return self
    
    def dict(self):
        return {'class' : self.__class__.__name__, 'parameters' : self.__dict__}
    
    def merged_dict(self,parameters=None):
        old = dict()
        return {'class':old['class'],'parameters':{**old['parameters'],**parameters}}
    
    def merge(self,parameters={}):
        adict = {'class' : self.__class__.__name__, 'parameters' : {**self.__dict__, **parameters}}
        #adict = merged_dict(parameters)
        return eval(adict['class'])(**adict['parameters'])
```

```python
from calendar import monthrange
import ipywidgets as widgets
style = {'description_width': '250px'}
layout = {'width': '600px'}
from IPython.display import display, clear_output

class VisWidget(BaseAction):
    def __init__(self, **kwargs):
        allowed_keys = {'timerange','customerId','host','instance','status'}
        int_keys = {'customerId'}
        string_keys = {'timerange','host','instance','status'}
        for key in allowed_keys:
            self.__dict__.update({key : None})
        self.__dict__.update((k, int(v)) for k, v in kwargs.items() if k in int_keys and len(str(v)) > 0 )    
        self.__dict__.update((k, v) for k, v in kwargs.items() if k in string_keys) 

class Time(object):         
    def __init__(self,year=None,month=None,day=None):
        self.year  = self._int_value(year)
        self.month = self._int_value(month)
        self.day   = self._int_value(day)
    
    def _int_value(self,value):
        if value == '' or value == None:
            return None
        return int(value)
        
class TimeRange(object):    
    def __init__(self,dataframe=None,year_from=None,month_from=None,day_from=None,year_to=None,month_to=None,day_to=None):
        if dataframe is None:
            self.start = Time(year=year_from,month=month_from,day=day_from).__dict__
            self.end   = Time(year_to,month_to,day_to).__dict__
        else:
            #print(dataframe)
            year_from = min(pd.unique(dataframe['year']))
            year_to = max(pd.unique(dataframe['year']))  
            month_from = min(pd.unique(dataframe[dataframe['year']== year_from]['month']))
            month_to = max(pd.unique(dataframe[dataframe['year']== year_to]['month']))
            day_from = min(pd.unique(dataframe[(dataframe['year']== year_from) & (dataframe['month'] == month_from)]['day']))
            day_to = max(pd.unique(dataframe[(dataframe['year']== year_to) & (dataframe['month'] == month_to)]['day']))
            self.start = Time(year_from,month_from,day_from).__dict__
            self.end   = Time(year_to,month_to,day_to).__dict__
            
        
def get_widget_timerange():
    return TimeRange(year_from=year_from.value, year_to=year_to.value, month_from=month_from.value, month_to=month_to.value).__dict__

def get_state():
    return VisWidget(timerange=get_widget_timerange(),customerId=customer_list.value,host=hosts_list.value,instance=instances_list.value,status=status_list.value)

def draw_graph(_res1):
    with out:
        clear_output()
        fig = plt.figure(figsize=(12,6))
        ax = fig.add_subplot(1,1,1)
        plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True) 
        line1, =plt.plot(_res1.date, _res1['outcome'], label='month')
        label(ax,30,80)
        plt.legend();
        plt.show(fig)
        #fig.savefig('/tmp/test.png')

def create_filled_timeseries(pfall2):
    _res = aggregate(pfall2,0,column='counter',groupby=['year','month','day']).sort_values(['date'])
    del _res['index']
    check(_res)
    _res = _res.sort_values(['date']).reset_index()  
    return _res

def output_graph():
    #clear_output()
    global _res
    _res = create_filled_timeseries(pfall2)
    draw_graph(_res)
        
def output_heatmap(month):
    #print(month)
    data2 = createData(pfall2,month, 'counter')
    piv = pd.pivot_table(data2, values="outcome",index=["hours"], columns=["days"], fill_value=0)
    
    _days = monthrange(int(year.value), month)[1]
    #piv.reindex(index=range(0,24), columns=range(1,_days + 1), fill_value=0)
    if len(piv) > 0:
        with out:
            #clear_output()
            #createHeatmap(piv, '',vmin=0,vmax=40)
            createHeatmap(piv, '')

def init(pfall1, customerId, host, instance, status):            
    pfall2 = pfall1[pfall1['customerId'] == customerId]      
    pfall2 = pfall2[(pfall2['host'] == host)]   
       
    if instance != 'all':
        pfall2 = pfall2[(pfall2['instance'] == instance)]    
      
    pfall2 = pfall2[(pfall2['status'] == status)]
    return pfall2


data_names = ['customerId','host','instance','status']
    
def init_lists(pfall1,customerId=None, key=None, names=data_names):
    global _old_state
    #_old_state = None
    index = names.index(key)
    
    pfall2 = pfall1
    
    if index == 0:
        customers = pd.unique(pfall1['customerId'])
        customer_list.options = customers
        if customerId != None:
            customer_list.value = customerId
        else:
            customer_list.value = customers[0]
            
    pfall2 = pfall2[pfall2['customerId'] == customer_list.value] 
            
    if index < 1:
        hosts = pd.unique(pfall2['host'])
        hosts_list.options     = np.append(hosts, 'all')
        hosts_value = hosts[0]
        hosts_list.value       = hosts_value
    pfall2 = pfall2[pfall2['host']==hosts_list.value]
    
    if index < 2:
        instances = pd.unique(pfall2['instance'])
        instances_list.options = np.append(instances, 'all')
        instance_value = instances[0]
        instances_list.value = instance_value
    pfall2 = pfall2[pfall2['instance']==instances_list.value]
    
    if index < 3:
        status = pd.unique(pfall2['status'])
        status_list.options = status 
        status_value = status[0]
        status_list.value = status_value
    pfall2 = pfall2[pfall2['status']==status_list.value]
     
    #_old_state = get_state()
    return pfall2
    
def get_value_split():
    return month.value.split('-')    
    
def is_state_updated():
    with out:
        clear_output()
    state = get_state()
    global _old_state
    if _old_state == None or state == _old_state:
        return False
    _old_state = None
    return True
    
def adapt(pfall1):  
    global pfall2
    state = get_state().__dict__
    print('adapt',state)
    pfall2 = pfall1[pfall1['customerId'] == state['customerId']]
    hosts_list.options     =  np.append( pd.unique(pfall2['host']), 'all')
    
    pfall2 = pfall2[(pfall2['host'] == state['host'])]
    instances_list.options = np.append(pd.unique(pfall2['instance']), 'all')
    
    if instances_list.value != 'all':
        pfall2 = pfall2[(pfall2['instance'] == state['instance'])]
    
    status_list.options = pd.unique(pfall2['status'])
    pfall2 = pfall2[(pfall2['status'] == state['status'])]
    
    print('adapt',get_state().__dict__)
    return pfall2
   

    
def adapt_date(pfall2):    
    state = get_widget_timerange()
    
    #if len(year.value) > 0:
    #    pfall2 = pfall2[(pfall2['year'] == int(year.value))]
    
    if  state['start']['month'] != None:
        if  state['end']['month'] == None:
            pfall2 = pfall2[(pfall2['month'] == state['start']['month'])]
            #output_heatmap(int(month.value))
        else:
            #del(pfall2['level_0'])
            _res=[]
            for m in range(state['start']['month'], state['end']['month'] + 1):
                #output_heatmap(m)
                _res.append(get_month(pfall2, year=state['start']['year'], month=m))
            pfall2 = pd.concat(_res, ignore_index=True).reset_index()
            #output_graph()
    # else:
    #     output_graph()   
    return pfall2 
    


def adapt_all(key):
    global _old_state
    global pfall2
    if is_state_updated():
        _old_state = None
        
        #print(_old_state.__dict__)
        #pfall2 = adapt(pfall1)
        pfall2 = init_lists(pfall1,customerId=customer_list.value, key=key, names=data_names)
        
        #pfall2 = adapt_date(pfall2)
        
        init_date_widget(pfall2)
        output_graph()
        #print(int(change.value))
        #print(len(pfall1),len(pfall2))
        _old_state=get_state()
    
def hosts_liston_change(change):
    global _old_state
    if change['type'] == 'change' and change['name'] == 'value':
        #print(change['owner'].description)
        adapt_all(change['owner'].description)
        
def on_value_submit_month(change):
    adapt_all()
```

```python
# https://machinelearningmastery.com/how-to-develop-lstm-models-for-time-series-forecasting/
# split a univariate sequence into samples

from numpy import array
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense

class LSTM_Filler(object):
    def __init__(self,_res):
        self.X, self.y = self.split_sequence(_res['outcome'], 30)
        self.create_model()
        self.fit()
        
    def split_sequence(self,sequence, n_steps):
        X, y = list(), list()
        for i in range(len(sequence)):
            # find the end of this pattern
            end_ix = i + n_steps
            # check if we are beyond the sequence
            if end_ix > len(sequence)-1:
                break
            # gather input and output parts of the pattern
            seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
            X.append(seq_x)
            y.append(seq_y)
        return np.array(X), np.array(y)

    def create_model(self):
        n_steps = 30

        # reshape from [samples, timesteps] into [samples, timesteps, features]
        n_features = 1
        self.X = self.X.reshape((self.X.shape[0], self.X.shape[1], n_features))
        # define model
        self.model = Sequential()
        self.model.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
        self.model.add(Dense(1))
        self.model.compile(optimizer='adam', loss='mse')
        
    def fit(self):   
        self.model.fit(self.X, self.y, epochs=200, verbose=0)

    def predict(self,x_input=None):    
        x_input = x_input.reshape((1, n_steps, n_features))
        yhat = model.predict(x_input, verbose=0)
        return yhat[0][0]
```
